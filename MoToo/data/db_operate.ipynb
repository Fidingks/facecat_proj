```python
import sqlite3

conn = sqlite3.connect('user.db')
cur = conn.cursor()

# 为SQLite启用外键约束
cur.execute("PRAGMA foreign_keys = ON;")

# 创建 user_info 表
cur.execute('''
CREATE TABLE IF NOT EXISTS user_info (
    user_id INTEGER UNIQUE PRIMARY KEY, 
    nick_name TEXT,
    add_time DATETIME,
    last_active_time DATETIME,
    invite_code TEXT,
    invited_by_code TEXT DEFAULT NULL,
    recharge_amount DECIMAL(10, 2) DEFAULT 0.00,
    invite_number INTEGER DEFAULT 0,
    invite_users_recharge DECIMAL(10, 2) DEFAULT 0.00, 
    register_points INTEGER DEFAULT 0,
    register_points_expiry_time DATETIME,
    invited_points INTEGER DEFAULT 0, 
    invited_points_expiry_time DATETIME,
    one_day_points INTEGER DEFAULT 0,
    one_day_points_expiry_time DATETIME,
    month_points1 INTEGER DEFAULT 0,
    month_points1_expiry_time DATETIME,
    month_points2 INTEGER DEFAULT 0,
    month_points2_expiry_time DATETIME,
    cupk_points INTEGER DEFAULT 0,
    cupk_points_expiry_time DATETIME
);
''')

# 创建 user_config 表并设置外键约束，使用 ON DELETE CASCADE
cur.execute('''
CREATE TABLE IF NOT EXISTS user_config (
    user_id INTEGER UNIQUE, 
    is_chat_with_image INTEGER DEFAULT 0,
    send_text_reply_voice INTEGER DEFAULT 0,
    send_voice_reply_text INTEGER DEFAULT 0,
    now_model TEXT DEFAULT 'gpt-4o-mini',
    now_voice TEXT DEFAULT 'alloy',
    now_prompt TEXT,
    voice_prompt TEXT,
    FOREIGN KEY (user_id) REFERENCES user_info(user_id) ON DELETE CASCADE
);
''')

# 创建 user_chat 表并设置外键约束，使用 ON DELETE CASCADE
cur.execute('''
CREATE TABLE IF NOT EXISTS user_chat (
    user_id INTEGER PRIMARY KEY,
    nick_name TEXT NOT NULL,
    chat_history TEXT,
    voice_history TEXT,
    image_history TEXT,
    chat_image1 TEXT,
    chat_image2 TEXT,
    chat_image3 TEXT,
    latest_image_time DATETIME,
    latest_file_time DATETIME,
    FOREIGN KEY (user_id) REFERENCES user_info(user_id) ON DELETE CASCADE
);
''')

conn.commit()
cur.close()
conn.close()
print("success")
```
